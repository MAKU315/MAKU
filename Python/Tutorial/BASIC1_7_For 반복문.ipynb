## 중첩 반복문

```python
for i in range(4):
    for j in range(4):
        print(i + j, end=" ")
    print()
```

```python
for i in range(6):
    n1 = i + 1
    for j in range(6):
        n2 = j + 1
        print(n1, n2)
```

```python
for i in range(6):
    D1 = i + 1
    for j in range(6):
        D2 = j + 2
        Sum = D1 + D2
        
        if (Sum % 4) == 0:
            print(D1, D2)

# range = [0 : 6]
```

```python
coin = 1024
for i in [1/2,2]:
    for j in [1/2,2]:
        for k in [1/2,2]:
            print(int(coin*i*j*k))
            
    
    
```

```python
n = 3

for a in range(1,11):
    
    for b in range(1, 11):
        
        for c in range(1, 11):
            if (a ** n + b **n) == (c ** n):
                #print( "a "+ str(a) + "  b "+ str(b) + "  c " + str(c))
                print( "a {}  b  {}  c  {}".format(a, b , c))
```

### 카운터 변수가 변하는 경우

```python
for j in range(10):
    sum = 0
    for i in range(j + 1):
        sum = sum + (i + 1)
    print(sum)
```

```python
for i in range(1,10):
    sum1 = 0
    sum2 = 0
    for j in range(1,i+1):
        sum2 = sum2 + j
        sum1 = sum1 + sum2
    print(sum1)
```
