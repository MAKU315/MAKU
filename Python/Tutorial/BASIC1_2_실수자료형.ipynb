### 이진법

### 컴퓨터는 이진법 - > 십진법으로 변환


#### 예를 들어 0.1은 이진법으로 무한 소수

```python
%precision 55
```

```python
0.1
```

```python
%precision %r
```

```python
0.1
```

### 실제 python 은 이진법을 더함 / R과 차이남

```python
0.1+0.2
```

### (0.1+0.2)는 0.3과 다름으로 round를 해야한다.

```python
(0.1+0.2) == 0.3
```

```python
round(0.1 + 0.2, 5) == round(0.3, 5)
```
