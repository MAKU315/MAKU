## 문자열 인코딩

* 문자는 2진 숫자(byte seq)
 + ASCII
 - CP949
 - 유니코드(UTF-8)

--------------------

### 한글 문자열의 분석을 위해서는 반드시 유니코드로 변환해야함

```python
u = "가"
print(u)
len(u)
```

```python
b = bytearray("가", 'cp949')
print(b)
len(b)
```

바이트열은 한글자씩 분리 가능

```python
u1 = "ABC"
u2 = "가나다"
print(len(u1), len(u2))
print(u1[0], u1[1], u1[2])
print(u2[0], u2[1], u2[2])
```

```python
b1 = bytearray("ABC", 'cp949')
b2 = bytearray("가나다", 'cp949')
print(len(b1), len(b2))
print(chr(b1[0]), chr(b1[1]), chr(b1[2]))
print(chr(b2[0]), chr(b2[1]), chr(b2[2]), chr(b2[3]))
```

## 인코딩과 디코딩

```python
u = "가나다"
type(u)
```

```python
b1 = u.encode("cp949")
print(b1)
type(b1)
```

```python
b1
```

```python
b2 = u.encode("euc-kr")
type(b2)
```

```python
b2
```

```python
b3 = u.encode("utf-8")
type(b3)
```

```python
b3
```

* 바이트를 디코딩 할때, 같은 방식을 사용

```python
b1.decode("cp949")
```

```python
b1.decode("euc-kr")
```

```python
b1.decode("utf-8")
```
