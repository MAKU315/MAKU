## if ~ elif ~ else 문

```python
score = 8

if score >= 8 :
    print("A"+"학점이다~!")
elif 5 <= score < 8 :
    print("B"+"학점이다..")
else :
    print("C"+"학점이다ㅜㅜ")
    
```

```python
grade = 8

if grade > 10 :
    print("1"+"등급이다~!")
elif 7 < grade:
    print("2"+"등급이다..")
elif 4 < grade:
    print("3"+"등급이다..;;")
else :
    print("4"+"등급이다ㅜㅜ")
    
```
