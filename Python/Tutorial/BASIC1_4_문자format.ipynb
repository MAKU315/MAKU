## 데이터 사이언스 스쿨(https://datascienceschool.net) 을 기초로 합니다.

### %를 활용

```python
"I'm %s" % "MAKU"
```

```python
"나는 %d 살 이다." % 3
```

### 여러 값을 넣어야 되면, (   ) 소괄호 사용

```python
"%.3f is the value of %s" % (3.14, "MAK")
```

```python
"%.3f is the value of %s" % ("MAK",3.14) ## 형식이 맞지 않으면 오류
```

## %를 활용 고급 문자 추가

```python
"[%20s]" % ("*"*10)
```

```python
"[%20s]" % "*"*10
```

```python
"[%-20s]" % "*"
```

```python
"[%20f]" % (33)
```

```python
"[%20d]" % (33)
```

```python
x = 1 / 3
"X는 %.3f" % x
```

## format을 활용한 문자 기입

```python
"I'm {}".format("MAKU")
```

```python
"{} is the value of {}".format(3.14, "MAK")
```

### format은 인덱싱 가능

```python
"{1}'s score is {0}, {0}, {0}".format(100, "Jane")
```

```python
"[{:20}]".format("*"*10) # 0 ~ 19 까지 만듬
```

```python
"[{:20}]".format("*")
```

### format 은  f, s, d 사용 가능

```python
"[{:<20.5f}]".format(x) # 
```

```python
"[{:>20.5f}]".format(x) # 
```

## 연습문제

```python
# 8 개
a = 3
b = 12

print("{:>8d}".format(a) +"\n" +"x"+"{:>7d}".format(b) + "\n" + "-"*8,"\n" +"{:>8d}".format(a * b))
```
