### 리스트

#### 인덱싱(Indexing) 가능 - 순서

```python
x = [88, 90, 100]
```

```python
x[0]
```

```python
score = [1,3,88, 90, 100]
(score[0]+score[1]+score[2]+score[3]+score[4])/len(score)
```

### 딕셔너리

#### 키(key) : 값(value)

```python
b = {"math": 88, "english": 90, "history": 100}
```

```python
b[0] # 순서형이 아니기 때문에 indexing 숫자로 불가
```

```python
b["math"]
```

```python
(b["math"] + b["english"]) / 2
```

```python
b2 = {1: 10, 2: 20}
```

```python
b2
```

```python
b # 키(key)의 순서로 정렬 됨
```

```python
day = {"01": 31, "02": 28, "03": 31, "04":30, "05":31, "06":30
    , "07":31, "08":31, "09":30, "10":31, "11":30, "12":31}
```

```python
(day["02"]+day["04"]+day["06"]+day["08"]+day["10"]+day["12"])/6
```
