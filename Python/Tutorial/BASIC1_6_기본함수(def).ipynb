## def == define

```python
def 음료(x):
    if x <= 500 :
        y = "생수"
    else :
        y = "콜라"
    return y

print(음료(500))
print(음료(1000))   
    
```

```python
def twotimes(x):
    y = 2 * x
    return y

twotimes(2)
```

```python
# 1.
def odd_fnt(a):
    if a % 2 == 0 :
        print("짝수")
    else :
        print("홀수")
        
odd_fnt(2)
```

```python
# 2. 
def year_fnt(y):
    if ((y % 4)==0) & (((y % 100)!=0) | ((y % 400)==0)) :
        print("윤년")
    else :
        print("평년")

year_fnt(2016)    
```

```python
# 3.
def days(month):
    if (month in [1,3,5,7,8,10,12]):
        day = 31
    elif month == 2:
        day = 28
    else :
        day = 30
    return day

days(11)
```
